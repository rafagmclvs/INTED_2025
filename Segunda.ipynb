```python
#8
dia= int(input("Digite o valor do dia: "))
match dia:
  case 1: print("Domingo")
  case 2: print('Segunda-feira')
  case 3: print('Terça-feira')
  case 4: print('Quarta-feira')
  case 5: print('Quinta-feira')
  case 6: print("Sexta-feira")
  case 7: print("Sábado")
  case _: print('Dia não encontrado')
```

```python
#10
qtde_trab = int(input('Digite as horas trabalhadas: '))
valor_hora = float(input("Digite o valor: "))
if qtde_trab <= 40:
  print('R$' ,qtde_trab * valor_hora)
else:
  print('R$' ,(qtde_trab * valor_hora)+(qtde_trab-40) * valor_hora * 1.5)
```
